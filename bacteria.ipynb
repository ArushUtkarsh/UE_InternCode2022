```python
import py4cytoscape as p4c
p4c.cytoscape_ping()
p4c.cytoscape_version_info()

# Import the class
import kmapper as km

# Initialize
mapper = km.KeplerMapper(verbose=1)

import pandas as pd
df = pd.read_csv('../Data/bacteria.csv')

print(df)

t = [[0 for i in range(135)] for j in range(217)]
x = [0 for i in range(217)]
    
import numpy as np

for i in range(217):
    x[i] = np.array(df.iloc[i].values)
    for j in range(134):
        t[i][j] = x[i][j+1]

u = np.array(t)
print(u)
```

```python
import sklearn as sk

# Fit to and transform the data
projected_data = mapper.fit_transform(u, projection=[2,3]) # X-Y axis

# Create a cover with 10 elements
cover = km.Cover(n_cubes=50)

# Create dictionary called 'graph' with nodes, edges and meta-information
graph = mapper.map(projected_data, u, cover=cover, clusterer = sk.cluster.DBSCAN(eps=0.5, min_samples=1))

print(graph)

# Visualize it
temp = mapper.visualize(graph, path_html="bacteria.sif",
                 title="bacteria")
```

```python
nodes = graph['nodes']
links = graph['links']
nodelist = nodes.keys()
source = []
target = []
for i in links.keys():
    ways = links[i]
    for j in ways:
        source.append(i)
        target.append(j)
nodess = pd.DataFrame(data={'id': nodelist})
edges = pd.DataFrame(data={'source': source, 'target': target})
p4c.create_network_from_data_frames(nodess, edges, title="bacteria", collection="Bacteria")
```
