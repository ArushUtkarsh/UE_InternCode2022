```python
import py4cytoscape as p4c
p4c.cytoscape_ping()
p4c.cytoscape_version_info()

# Import the class
import kmapper as km

# Initialize
mapper = km.KeplerMapper(verbose=1)

import pandas as pd
df = pd.read_csv('../Data/bacteria.csv')

df2 = df.copy(deep = True)
df2.drop('PatientID', inplace = True, axis = 1)
df2=df2.div(df2.sum(axis=1), axis=0)
nsel=df2>0.01
nsel=nsel.sum(axis=0)
x=nsel[nsel<10].index #5%
df2.drop(columns=x,inplace=True)
df2=df2.div(df2.sum(axis=1), axis=0) #Renormalize

df2withPatient = df2.copy(deep = True)
df2withPatient['PatientID'] = df['PatientID']

labeldf = pd.read_csv('../Data/labels.csv')
cluster1 = labeldf[labeldf['labels'] == 1]
cluster2 = labeldf[labeldf['labels'] == 2]
cluster3 = labeldf[labeldf['labels'] == 3]

temp = cluster1['PatientID'].tolist()
cluster1df = pd.DataFrame()
for i in df['PatientID']:
    if i in temp:
        cluster1df = cluster1df.append(df2withPatient.loc[df2withPatient['PatientID'] == i])
cluster1df.drop('PatientID', inplace = True, axis = 1)
print(cluster1df)
print()
        
temp = cluster2['PatientID'].tolist()
cluster2df = pd.DataFrame()
for i in df['PatientID']:
    if i in temp:
        cluster2df = cluster2df.append(df2withPatient.loc[df2withPatient['PatientID'] == i])
cluster2df.drop('PatientID', inplace = True, axis = 1)
print(cluster2df)
print()
        
temp = cluster3['PatientID'].tolist()
cluster3df = pd.DataFrame()
for i in df['PatientID']:
    if i in temp:
        cluster3df = cluster3df.append(df2withPatient.loc[df2withPatient['PatientID'] == i])
cluster3df.drop('PatientID', inplace = True, axis = 1)
print(cluster3df)
print()


t = [[0 for i in range(115)] for j in range(23)]
x = [0 for i in range(23)]
    
import numpy as np

for i in range(23):
    x[i] = np.array(cluster1df.iloc[:,i].values)
    for j in range(115):
        t[i][j] = x[i][j]

u = np.array(t)

t = [[0 for i in range(88)] for j in range(23)]
x = [0 for i in range(23)]

for i in range(23):
    x[i] = np.array(cluster1df.iloc[:,i].values)
    for j in range(88):
        t[i][j] = x[i][j]
        
v = np.array(t)
        
t = [[0 for i in range(14)] for j in range(23)]
x = [0 for i in range(23)]

for i in range(23):
    x[i] = np.array(cluster1df.iloc[:,i].values)
    for j in range(14):
        t[i][j] = x[i][j]
        
w = np.array(t)
```

```python
import sklearn as sk
from sklearn import ensemble

# Fit to and transform the data
projected_data = mapper.fit_transform(u, projection="l2norm")
projected_data2 = mapper.fit_transform(v, projection="l2norm")
projected_data3 = mapper.fit_transform(w, projection="l2norm")

# Create a cover with 10 elements
cover = km.Cover(n_cubes=10, perc_overlap = 0.7)

from scipy.special import comb
import scipy.misc
scipy.misc.comb = comb
import ecopy as ep
shannonH = ep.diversity(u, 'shannon')
shannonH2 = ep.diversity(v, 'shannon')
shannonH3 = ep.diversity(w, 'shannon')

# Create a custom 1-D lens with Isolation Forest
model = ensemble.IsolationForest(random_state=1729)
model.fit(u)
lens1 = model.decision_function(u).reshape((u.shape[0], 1))
model.fit(v)
lens2 = model.decision_function(v).reshape((v.shape[0], 1))
model.fit(w)
lens3 = model.decision_function(w).reshape((w.shape[0], 1))

from sklearn.decomposition import PCA
pca = PCA()
pca.fit(u)
#print(pca.explained_variance_r_atio_)

#temp = np.array([pca.explained_variance_ratio_[1]])
#arr = [temp]
#for i in range(134):
    #arr = np.concatenate((arr,[np.array([pca.explained_variance_ratio_[i+1]])]), axis=1)

# Combine lenses pairwise to get a 2-D lens i.e. [Isolation Forest, L^2-Norm] lens
#lens = np.c_[pca.explained_variance_ratio_, shannonH]
#lens = np.c_[projected_data, shannonH]
lens = np.c_[projected_data3,shannonH3]

# Create dictionary called 'graph' with nodes, edges and meta-information
# graph = mapper.map(projected_data, u, cover=cover, clusterer = sk.cluster.DBSCAN(eps=0.1, min_samples=1))
graph = mapper.map(lens, w, cover=cover, clusterer = sk.cluster.AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='single'))

print(graph)

# Visualize it
temp = mapper.visualize(graph, path_html="bacteria.sif", title="bacteria")
```

```python
for i in graph['nodes']:
    for j in range(len(graph['nodes'][i])):
        graph['nodes'][i][j] = df2.columns[graph['nodes'][i][j]]
        
for i in graph['links']:
    stack = []
    temp2 = graph['nodes'][i]
    for j in temp2:
        for k in graph['nodes']:
            if k!= i:
                if j in graph['nodes'][k]:
                    stack.append(k)
    graph['links'][i] = stack

for i in graph['nodes']:
    print(i,": ",graph['nodes'][i])
    
stack = []
for i in graph['nodes']:
    stack.append(i)

df3 = pd.DataFrame(stack, columns=['name'])

stack = []
for i in graph['nodes']:
    stack.append(graph['nodes'][i])
df3['#AllMembers'] = stack

bacteria = df2.columns.values.tolist()
print(bacteria)

for i in bacteria:
    stack = []
    for j in graph['nodes']:
        if i in graph['nodes'][j]:
            stack.append(1)
        else:
            stack.append(0)
    df3[i] = stack

rows = df3.shape[0]
cols = df3.shape[1]

for i in range(cols-2):
    stack = []
    for j in range(rows):
        count = 0
        flag = 0
        if df3[bacteria[i]][j] == 1:
            flag = 1
        for k in range(cols-2):
            if df3[bacteria[k]][j] == 1:
                count = count + 1
        val = flag/count
        stack.append(val)
    df3['conc_'+bacteria[i]] = stack

#tempdict = {}
#for i in graph['nodes']:
    #str = graph['nodes'][i][0]
    #for j in range(len(graph['nodes'][i])-1):
        #str+=', '+graph['nodes'][i][j+1]
    #tempdict[str] = graph['nodes'][i]
    
#graph['nodes'] = tempdict
#print(graph)
# Visualize it
```
